```python
str=input("enter string:")
covow=0
cocon=0
for i in str:
    if i in 'aeiouAEIOU':
        covow+=1
    else:
        cocon+=1
print("vowel=",covow)
print("consonant=",cocon)
```

```python
#Count uppercase and lowercase letters
s=input("string: ")
u=0
l=0
for i in s:
    if i.isupper():
        u += 1
    elif i.islower():
        l += 1
print("upper =", u)
print("lower =", l)
```

```python
#Replace all vowels with *
s = input("enter string: ")
result = ""
for i in s:
    if i in "aeiouAEIOU":
        result += "*"
    else:
        result += i
print(result)
```

```python
#match case just like switch 
day =int(input("Enter day: "))
match day:
    case 1:
        print("Monday")
    case 2:
        print("Tuesday")
    case 3:
        print("Wednesday")
    case _:
        print("Invalid")
```

```python
num=int(input("enter no.:"))
match num:
    case 1|3|5:
        print("odd")
    case 2|4|6:
        print("even")
    case _:
        print("range finish")
```

```python
x=int(input("enter no.:"))
match x:
    case x if x>0:
        print("+ve")
    case x if x<0:
        print("-ve")
    case _:
        print("0")
```
