```python
present = ["vipin", "rahul", "aman"]
absent = ["rahul", "rohit"]
present_set = set(present)
absent_set = set(absent)
common = present_set & absent_set
only_present = present_set - absent_set
only_absent = absent_set - present_set
attendance = {
    "common": sorted(list(common)),
    "present_only": sorted(list(only_present)),
    "absent_only": sorted(list(only_absent))
}
print("Attendance Analysis")
print("Students marked in both lists:", ", ".join(attendance["common"]) if attendance["common"] else "None")
print("Students only present:", ", ".join(attendance["present_only"]) if attendance["present_only"] else "None")
print("Students only absent:", ", ".join(attendance["absent_only"]) if attendance["absent_only"] else "None")
```
