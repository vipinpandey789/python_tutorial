```python
#Count Even & Odd (Function + List)
def evod():
    num=list(map(int,input("enter elements:" ).split()))
    ce=0
    co=0
    for i in num:
        if i%2==0:
            ce+=1
        else:
            co+=1
    return ce,co
even, odd = evod()
print("Even count:", even)
print("Odd count:", odd)
```

```python
#Largest & Second Largest (IMPORTANT 🔥)
def larsec():
    num=list(map(int,input("enter elements:").split()))
    l1=float('-inf')
    l2=float('-inf')
    for i in num:
        if i >l1:
            l2=l1
            l1=i
        elif i>l2 and i!=l1:
            l2=i
    return l1,l2
largest,second=larsec()
print("Largest:", largest)
print("Second Largest:", second)
```

```python
#Count Vowels in List of Strings
def vowel():
    s=(input("enter elements:").split())
    count=0
    for word in s:
        for ch in word: 
            if ch in 'AEIOUaeiou':
                count+=1
    return count
print(vowel())
```

```python
#Function Calling Function
def square(n):
    return [i*i for i in n]
def cube(n):
    return[i*i*i for i in n]
def total(n):
    sq=square(n)
    cu=cube(n)
    result=[]
    for i in range(len(n)):
        result.append(sq[i]+cu[i])
    return result
n=list(map(int,input("enter element:").split()))
print(total(n))        
```

```python
'''Mixed Challenge (REAL LOGIC)
square each number
keep only even squares
find sum'''
def mix():
    num=list(map(int,input("enter element:").split()))
    square=[i*i for i in num]
    print("square=",square)
    even=[i for i in square if i%2==0]
    print("even=",even)
    total=0
    for i in even:
        total+=i
    print("sum=",total)
    return total
mix()
    
```
