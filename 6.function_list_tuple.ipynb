```python
#functions with parameters
def greet(name):   #name gets value when function is called
    print("hello",name)   
greet('vipin')
```

```python
#multiple parameters
def add(a,b,c):
    addition=a+b+c
    return addition

print(add(1,2,3))
```

```python
#return value 
def add(a,b):
    return a+b
c=add(5,6)
print(c)
```

```python
#Even or Odd (Return Concept)
def check_even_odd(num):
    if num%2==0:
        return "even"
    else:
        return "odd"
print(check_even_odd(7))
print(check_even_odd(4))
```

```python
#sum of 1 to n 
def sum_n(n):
    s=0
    for i in range(1,n+1):
        s=s+i
    return s
print(sum_n(6))
```

```python
# multiple return values 
def cal(n):
    square=n*n
    cube=n*n*n
    return square,cube
print(cal(5))
```
