```python
def kbc_game():

    questions = [
        ("What is the extension of Python files?", [".py", ".java", ".txt", ".doc"], "A"),
        ("Which keyword is used to define a function?", ["func", "define", "def", "function"], "C"),
        ("What will print(type(5)) output?", ["<class 'int'>", "<class 'str'>", "<class 'float'>", "<class 'bool'>"], "A"),
        ("Which data type stores multiple values?", ["int", "float", "list", "char"], "C"),
        ("What is 2 + 3 * 2?", ["10", "8", "12", "6"], "B"),
        ("Which keyword is used for conditions?", ["loop", "if", "check", "case"], "B"),
        ("len('Python') returns?", ["5", "6", "7", "Error"], "B"),
        ("Which function takes input?", ["get()", "scan()", "input()", "read()"], "C"),
        ("What is 10 // 3?", ["3.3", "3", "4", "2"], "B"),
        ("Which is immutable?", ["list", "set", "tuple", "dict"], "C"),
        ("bool(0) gives?", ["True", "False", "Error", "None"], "B"),
        ("Which loop is used for fixed iterations?", ["while", "loop", "for", "do"], "C"),
        ("range(5) gives?", ["1-5", "0-5", "0-4", "Error"], "C"),
        ("'Hello'[1] gives?", ["H", "e", "l", "o"], "B"),
        ("Which keyword handles exceptions?", ["catch", "try", "except", "error"], "B"),
        ("type([]) gives?", ["list", "tuple", "dict", "set"], "A"),
        ("2 ** 3 gives?", ["6", "8", "9", "5"], "B"),
        ("break does what?", ["skip", "stop loop", "pause", "restart"], "B"),
        ("'5' + '5' gives?", ["10", "55", "Error", "None"], "B"),
        ("Which method adds to list?", ["add()", "insert()", "append()", "push()"], "C")
    ]

    prize_money = [
        1000, 2000, 3000, 5000, 10000,
        20000, 40000, 80000, 160000, 320000,
        640000, 1250000, 2500000, 5000000, 10000000,
        20000000, 30000000, 50000000, 70000000, 100000000
    ]

    total_won = 0

    print("🎮 Welcome to Python KBC Game 🎮")

    for i in range(len(questions)):
        q = questions[i][0]
        options = questions[i][1]
        correct = questions[i][2]

        print("\n-----------------------------")
        print(f"Question {i+1} for ₹{prize_money[i]}")
        print(q)

        print("A.", options[0])
        print("B.", options[1])
        print("C.", options[2])
        print("D.", options[3])

        answer = input("Enter your answer (A/B/C/D): ").upper()

        if answer == correct:
            total_won = prize_money[i]
            print("✅ Correct Answer!")
            print(f"💰 Total Winning: ₹{total_won}")
        else:
            print("❌ Wrong Answer!")
            print(f"🏁 Game Over! You won ₹{total_won}")
            break

    else:
        print("\n🎉 Congratulations! You answered all questions correctly!")
        print(f"🏆 You won ₹{total_won}")


# Run the game
kbc_game()
```
