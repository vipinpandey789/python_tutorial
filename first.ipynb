```python
#Check if a number is prime
num=int(input("enter no. to check="))
if(num<2):
    print("not prime")
else:
    limit=int(num ** 0.5)
    for i in range(2,limit+1):
        if num%i==0:
            print("not a prime")
            break
    else:
        print("prime")
```

```python
#or prime
num = int(input("Enter number: "))
if num < 2:
    print("Not prime")
else:
    for i in range(2, num):
        if num % i == 0:
            print("Not prime")
            break
    else:
        print("Prime")
```
