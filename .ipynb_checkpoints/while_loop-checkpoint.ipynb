```python
#Print 1 to 10
i=1
while i<=10:
    print(i,end=",")
    i+=1
```

```python
#Print even numbers
i = 2
while i <= 20:
    print(i,end=",")
    i += 2
```

```python
#Reverse numbers
i = 10
while i >= 1:
    print(i,end=",")
    i -= 1
```

```python
#Find sum of numbers from 1 to 10
s=0
i=1
while i<=10:
    s=s+i
    i+=1
print(s)
```

```python
#Count how many numbers are divisible by 3 (1 to 30)
i=1
while i<=30:
    if i%3==0 and i%5==0:
        print(i)
    i+=1
```

```python
#reverse any no.
n=int(input("enter no. to reverse:"))
rev=0
while n>0:
    r=n%10
    rev=rev*10+r
    n=n//10
print(rev)
```

```python
#Check if number is palindrome
n=int(input("enter no. to check:"))
temp=n
rev=0
while n>0:
    r=n%10
    rev=rev*10+r
    n=n//10
print(rev)
if(rev==temp):
    print("yes")
else:
    print("no")
    
```

```python
#Keep taking input until user enters 0
n=int(input("enter user's choice:"))
while n>=0:
    if n==0:
        print(n)
        break
    else:
        n=int(input("enter user's choice:"))
```

```python
#Print Fibonacci series (first n terms)
n=int(input("enter range:"))
a=0
b=1
i=0
while i<n:
    print(a,end=" ")
    c=a+b
    a=b
    b=c
    i+=1
    
    
```
