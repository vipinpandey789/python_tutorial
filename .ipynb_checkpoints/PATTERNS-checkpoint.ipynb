```python
for i in range(9):
    for j in range(9):
        if(i==j or i+j==8):
            print("*",end="")
        else:
            print(" ",end="")
    print()
```

```python
for i in range(6):
    for j in range(i):
        print("*",end="")
    print()
```

```python
i=1
while i<=5:
    j=1
    while j<=i:
        print("*",end="")
        j+=1
    print()
    i+=1
```
