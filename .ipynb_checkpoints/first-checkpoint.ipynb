```python
#match case
x=4
match x:
    case 0:
        print("x is 0")
    case 4:
        print("case is 4")
    case _:
        print(x)
```

```python
#Find sum of digits of a number
n = int(input("Enter number: "))
s = 0
for _ in range(len(str(n))):
    digit = n % 10
    s = s + digit
    n = n // 10
print("Sum of digits =", s)
```

```python
#Count vowels in a string
s=input("enter string=")
count=0
v='aeiouAEIOU'
for i in s:
    if i in v:
        count+=1
print("total=",count)
```
